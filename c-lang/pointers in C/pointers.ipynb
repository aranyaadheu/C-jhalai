## Definition: 

A pointer is simply a variable that stores a memory address rather than a value.

- Normal Variable (```int a = 10;```): Stores the value ```10```.

- Pointer Variable (```int *p;```): Stores the hex address (e.g., ```0x7ffd```) where 10 lives in the RAM.

It is just a signpost pointing to a location.

## The Syntax:

only two operators. 

1. ```&``` (Address-of Operator: "Get me the address of this variable.")

2. ```*``` (Dereference Operator): "Go to this address and give me the value stored there."
