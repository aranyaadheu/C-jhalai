# What is a Function?

- Functions are used to divide a large c program into smaller pieces.

- A function can be called multiple ties to provide resuability and modualrity to the C program

- Also called procedure or subroutine

# What is a Recursive Function?

- Recursive functions or recursion is a process when a function calls a copy of itself to work on smaller problem.

- Any function which calls itself is called recursive function

- This make the life of a programmer easy by dividing a given problem into easier one. 

- A termination condition is imposed on such sfunctions to stop them executing copies of themselves forever. 

- Any problem that can be solve recursively, can also be solved iteratively.

# Ex 1: Factorial Calculation. 

- the case at which the function doesn't recur is called the based case. 

- The instances where the function keeps calling itself to perform a subtask, is called the recursive case. 

- for factorial calculation the base case occurs at the parameter value of 0 and 1. 

```bash
n! = n × (n−1)!
0! = 1
```

```c
#include <stdio.h>

int factorial(int n)
{
    if (n == 0)        // base case
        return 1;
    else
        return n * factorial(n - 1);  // recursive call
}

int main()
{
    printf("%d", factorial(5));
    return 0;
}
```
How it works (factorial(5))

```bash
5 * factorial(4)
4 * factorial(3)
3 * factorial(2)
2 * factorial(1)
1 * factorial(0) → 1
```
