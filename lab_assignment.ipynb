### Problem 2: Write a program that takes an integer, a float, and a character as input and prints them back with their respective sizes in bytes using the ```sizeof()``` operator.
Solution: 

```c
#include <stdio.h>

int main() {
    int i;
    float f;
    char c;

    // Simplified input
    printf("Enter an int, a float, and a char: ");
    scanf("%d %f %c", &i, &f, &c);

    // Displaying values and sizes
    printf("\nValue: %d, Size: %zu bytes", i, sizeof(i));
    printf("\nValue: %f, Size: %zu bytes", f, sizeof(f));
    printf("\nValue: %c, Size: %zu bytes", c, sizeof(c));

    return 0;
}

```

Output:

```bash
Enter an int, a float, and a char: 10 5.5 A

Value: 10, Size: 4 bytes
Value: 5.500000, Size: 4 bytes
Value: A, Size: 1 bytes
```

# Problem 1: Create a program to calculate the area of a circle where the value of PI is defined as a ```const```.
Solution:

```c
#include <stdio.h>

int main() {
    // Defining PI as a constant
    const float PI = 3.14159;
    float radius, area;

    printf("Enter the radius of the circle: ");
    scanf("%f", &radius);

    // Area calculation
    area = PI * radius * radius;

    printf("The area of the circle is: %.2f\n", area);

    return 0;
}
```

Output: 

```bash
Enter the radius of the circle: 5
The area of the circle is: 78.54
```
